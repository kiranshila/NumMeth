### Deliverable 1 - estimate_gaussian_params

```julia
function estimate_gaussian_params(data::Vector)
  N = length(data)
  mean = (1/N)*sum(data)
  # Using the modified variance function
  variance = (1/(N-1))*(sum((data .- mean).^2) - (1/N)*(sum(data .- mean))^2)
  std = sqrt(variance)
  skew = (1/N)*sum(((data .- mean)./std).^3)
  kurt = ((1/N)*sum(((data .- mean)./std).^4))-3

  return Dict("Mean"=>mean,"Variance"=>variance,"Std. Deviation"=>std,"Skew"=>skew,"Kurtosis"=>kurt)
end
```

```julia
# Find Stats for dat
stats = estimate_gaussian_params(dat)
```

### Deliverable 3 - kernel_density_estimator

So, I'll start by defining the kernel functions.

```julia
function epanechnikov(t::Number,h::Number)
  if abs(t) < h
    return 0.75 * (1 - (t/h)^2) / h
  else
    return 0
  end
end
```

```julia
function gaussian(t::Number,h::Number)
  return 1/(sqrt(pi*2)*h) * exp(-t^2/(2*h^2))
end
```

```julia
function rectangular(t::Number,h::Number)
  if abs(t) < h
    return 1/(2*h)
  else
    return 0
  end
end
```

And now the actual kernel density estimation function.

```julia
function kernel_density_estimator(data::Vector,bw::Number,numPoints::Int64=200,kernel::Function=epanechnikov)
    @assert bw > 0 "Bandwidth must be positive"
    return  range(minimum(data),stop=maximum(data),length=numPoints),
            [(1/length(data)) * sum([kernel(x-xi,bw) for xi in data]) for x = range(minimum(data),
                                                                           stop=maximum(data),
                                                                           length=numPoints)]
end
```

### Deliverable 4 - Testing and Validation

So now lets plot some histograms with their kernel density estimations.

#### Epanechnikov

```julia
estimation = kernel_density_estimator(dat,0.2)
plt = histogram(dat,ylabel="Count",color=:gold,legend=false,grid=false)
plot!(twinx(),estimation[1],estimation[2],color=:green,legend=true,grid=false,label="Kernel Density")
```

And now a different bandwith

```julia
estimation = kernel_density_estimator(dat,0.05)
histogram(dat,ylabel="Count",color=:gold,legend=false)
plot!(twinx(),estimation[1],estimation[2],color=:green,legend=true,grid=false,label="Kernel Density")
```

Certainly noisier, how about a high bandwith?

```julia
estimation = kernel_density_estimator(dat,1)
histogram(dat,ylabel="Count",color=:gold,legend=false)
plot!(twinx(),estimation[1],estimation[2],color=:green,legend=true,grid=false,label="Kernel Density")
```

Now we lost the two peaks. There certainly is a tradeoff.

Lets look at the other windowing functions with the 0.2 bandwidth.

#### Rectangular

```julia
estimation = kernel_density_estimator(dat,0.2,200,rectangular)
histogram(dat,ylabel="Count",color=:gold,legend=false)
plot!(twinx(),estimation[1],estimation[2],color=:green,legend=true,grid=false,label="Kernel Density")
```

Obviously flat now, but noisier than the epanechikov.

```julia
estimation = kernel_density_estimator(dat,0.4,200,rectangular)
histogram(dat,ylabel="Count",color=:gold,legend=false)
plot!(twinx(),estimation[1],estimation[2],color=:green,legend=true,grid=false,label="Kernel Density")
```

Even raising the bandwith started flattening it out before it got less noisy. How about gaussian?

#### Gaussian

```julia
estimation = kernel_density_estimator(dat,0.2,200,gaussian)
histogram(dat,ylabel="Count",color=:gold,legend=false)
plot!(twinx(),estimation[1],estimation[2],color=:green,legend=true,grid=false,label="Kernel Density")
```

The middle dip isn't as low as the equivalent epanechnikov.
